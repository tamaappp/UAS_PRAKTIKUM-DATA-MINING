```python
# Mengimpor pustaka yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Memuat dataset Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Membuat model K-Means dengan 3 cluster
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
# Melatih model menggunakan data
kmeans.fit(X)
```

```python
# Mendapatkan label hasil clustering
labels = kmeans.labels_
```

```python
# Menggunakan PCA untuk mengurangi dimensi menjadi 2D untuk visualisasi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Plot hasil clustering
plt.figure(figsize=(10, 5))
```

```python
# Plot data asli
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('Data Asli')
plt.xlabel('Komponen Utama 1')
plt.ylabel('Komponen Utama 2')
```

```python
# Plot hasil clustering
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', edgecolor='k')
plt.title('Hasil Clustering dengan K-Means')
plt.xlabel('Komponen Utama 1')
plt.ylabel('Komponen Utama 2')

plt.tight_layout()
plt.show()
```

```python
# Menampilkan hasil clustering dan label asli untuk beberapa data
print("Hasil Clustering (K-Means)\tLabel Asli")
for label, actual in zip(labels, y):
    print(f"{label}\t\t\t{actual}")
```
